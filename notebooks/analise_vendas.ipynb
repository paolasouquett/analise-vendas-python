üìå Observa√ß√£o: os valores monet√°rios (`preco_unitario` e `faturamento`)
est√£o expressos em Reais (R$). As quantidades representam unidades vendidas.





```python
import pandas as pd

# Carregar o dataset
df = pd.read_csv('../data/vendas.csv')

# Visualizar as primeiras linhas
df.head()
```

```python
# Criar coluna de faturamento
df['faturamento'] = df['quantidade'] * df['preco_unitario']

# Verificar
df.head()
```

## üìå KPIs Principais

A seguir, s√£o apresentados os principais indicadores de desempenho
extra√≠dos a partir dos dados de vendas.


```python
# Faturamento total
faturamento_total = df['faturamento'].sum()

# Ticket m√©dio
ticket_medio = df['faturamento'].mean()

print(f"Faturamento total: R$ {faturamento_total:.2f}")
print(f"Ticket m√©dio: R$ {ticket_medio:.2f}")

```

## üìà An√°lises Explorat√≥rias

Nesta se√ß√£o, analisamos o desempenho de vendas por categoria e
produto.



```python
faturamento_categoria = (
    df.groupby('categoria')['faturamento']
    .sum()
    .sort_values(ascending=False)
)

faturamento_categoria
```

```python
produtos_mais_vendidos = (
    df.groupby('produto')['quantidade']
    .sum()
    .sort_values(ascending=False)
)

produtos_mais_vendidos
```

## üåç An√°lise de Vendas por Regi√£o

Nesta se√ß√£o, analisamos o faturamento e o volume de vendas por regi√£o
para identificar onde est√£o as maiores oportunidades de neg√≥cio.


```python
# Quantidade total vendida por regi√£o
quantidade_regiao = (
    df.groupby('regiao')['quantidade']
    .sum()
    .sort_values(ascending=False)
)

quantidade_regiao
```

```python
# Faturamento total por regi√£o
faturamento_regiao_formatado = faturamento_regiao.map(
    lambda x: f"R$ {x:,.2f}"
)

faturamento_regiao_formatado


```

De forma geral, os resultados demonstram como a an√°lise de dados pode

```python
# Produtos mais vendidos (quantidade)
produtos_mais_vendidos = (
    df.groupby('produto')['quantidade']
    .sum()
    .sort_values(ascending=True)
)

plt.figure()
produtos_mais_vendidos.plot(kind='barh')
plt.title('Produtos Mais Vendidos (Unidades)')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Produto')
plt.tight_layout()
plt.show()
```
